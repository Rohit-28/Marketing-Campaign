```python
#Create new KPIs in table
#CTR = Click through rate
#CR  = Conversion Rate
#CPCl = Cost per Click
#CPCNV = Cost per Conversion
social_media_data['CTR'] = (social_media_data.Clicks*100)/social_media_data.Impressions
social_media_data['CR'] = (social_media_data.Total_Conversion*100)/social_media_data.Clicks
social_media_data['CPCL'] = (social_media_data.Spent*100)/social_media_data.Clicks
social_media_data['CPCNV'] = (social_media_data.Spent*100)/social_media_data.Total_Conversion
```

```python
#New Data KPI generation, Revenue over Advertising Spendings(ROAS)
#New Data KPI generation, Cost per Million (CPMill)
#Assumptions:
#Business Specification 
#value/Total_conversion = $5
#value/Approved_Conversion = $100 
social_media_data['ROAS'] = ((social_media_data['Total_Conversion']*5+social_media_data['Approved_Conversion']*100)*100)/social_media_data.Spent
social_media_data['ROAS'] = social_media_data['ROAS'].apply(lambda x: round(x, 3))
social_media_data['CPMill'] = (social_media_data['Spent']/social_media_data['Impressions'])*1000
social_media_data['CPMill'] = social_media_data['CPMill'].apply(lambda x: round(x, 3))
social_media_data['LOG(ROAS)'] = social_media_data['ROAS'].apply(lambda x: algebra.log10(x))
```
